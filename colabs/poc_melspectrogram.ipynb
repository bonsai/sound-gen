# 🎵 親子対話：メルスペクトログラム — POC

メルスペクトログラムは音声AIの共通言語。音声波形を「料理のレシピ図」に変換する。

```
# サンプル音声ダウンロード
import urllib.request
url = "https://github.com/bonsai/sound-gen/raw/main/colabs/samples/speech_sample.wav"
try:
    urllib.request.urlretrieve(url, "sample.wav")
    print("✅ Sample downloaded")
except:
    print("⚠️  Download failed, generating synthetic audio")
    import numpy as np
    sr = 22050
    t = np.linspace(0, 3, sr*3)
    from scipy.io.wavfile import write
    audio = np.sin(2*np.pi*440*t) * (1 - t/3)
    write("sample.wav", sr, (audio * 32767).astype(np.int16))
```

```
# 波形 → メルスペクトログラム
import librosa
import matplotlib.pyplot as plt
import numpy as np

y, sr = librosa.load("sample.wav", sr=22050)

# 3種類のスペクトログラムを比較
fig, axes = plt.subplots(3, 1, figsize=(12, 8))

# 1. 通常のスペクトログラム
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='hz', ax=axes[0])
axes[0].set_title('① 通常スペクトログラム (線形周波数)')

# 2. メルスペクトログラム
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
S_db = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='mel', ax=axes[1])
axes[1].set_title('② メルスペクトログラム (メル尺度)')

# 3. MFCC
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
librosa.display.specshow(mfcc, sr=sr, x_axis='time', ax=axes[2])
axes[2].set_title('③ MFCC (メル周波数ケプストラム係数)')

plt.tight_layout()
plt.show()
```

---
### 📝 まとめ

- **通常STFT**: 高周波が細かくて見にくい
- **メルスペクトログラム**: 人間の聴覚特性に合わせて周波数を丸める → 音声AIの共通表現
- **MFCC**: さらに圧縮して13次元 → 音声認識の伝統的な入力
